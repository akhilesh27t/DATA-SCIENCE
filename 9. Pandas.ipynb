##### Data series :

   - It is a one dimensional array or list
   - It is having only one row or only one columns
   - It can have the data of any types(int,float,string etc)
   - Data series is always associated with the index
   - In Data series we can give the index of various types(int,string,datatype etc)
   - To create the data series we used the function pd.series()--- for creating series
   - If w are not proving the index it will by default start from 0
     
#####  Q. If we have numpy array then why we have series data ?

   - In case of numpy it will take by default integer as a index
   - But in series data we can take any thing asa index (int,string,datetime etc)
     
##### Q. How dataseries Stored in memory ?

   - The series Data is stored in the memory as a numpy array (its memory efficient)
   - The indexes are stored differently in the memory and there will be mapping between indexes and memory
