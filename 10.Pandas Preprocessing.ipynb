#### @ Handling the missing values :

   - Missing value:na,nan,no value 
   - If we have a missing value in the datset the ML model will not do the proper prediction
   - So there is need of handling the missing values

#### 1. Drop the missing value:

  - Syntax : dataframe.dropna()
  - It will iterate over the row wherever it found the missing value in the column it will simply delete that row.
  - axis=0--by default
  - axis=1:columns
  - how=any/all
  - By default its any
  - Whenever we are using multiple columns in the subset arguments it will perform or operation
