### Data Visualization:

   - to understand the data in a better way
   - plotting is the most important thing in the data visualization.
   - Data visualization library 

     1. Matplotlib
     2. seaborn
     3. plotly
     4. d3.js

#####  1. Matplotlib:
   - it is a core data visualization library in python
   - remaining all the libraries are built on the top of matplotlib
   - Matplotlib is a kind of abstraction to other library
       
         * Syntax: matplotlib. pyplot as plt
   - Matplotlib: Data visualization library name
   - pyplot: it is a submodule present in the matplotlib: it is using the core functionality of matlab for a plotting
   - plt-----as alias

#####  2. seaborn:

   - it is also data visualization library which is built on the top of matplotlib
   - seaborn is kind of advanced version of matplotlib
   - so whatever the things matplotlib is doing same thing seaborn will do in a better way

         Syntax: import seaborn as sns

   - when we are constructing the line plot using matplotlib and if we are passing single axis measures then by default it is considered as y-axis                      measures, and x-axis measures is generated by matplotlib(it is starting from 0 becoz list index starts from 0)
