### @ Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\magar\\Downloads\\Airbnb_Open_Data.csv")
df
```

```python
df.info()
```

```python
#1.Id columns:
df["id"].value_counts()
```

```python
df['id'].duplicated().sum()
```

```python
df.dropna(subset=["id"],inplace=True)
df.drop_duplicates(subset='id',inplace=True)
```

```python
df
```

```python
# 2.Name column
import re
df['NAME'] = df['NAME'].astype(str).str.strip()
df['NAME'].replace('', pd.NA, inplace=True)
df['NAME'] = df['NAME'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', x))  
df['NAME'] = df['NAME'].apply(lambda x: re.sub(r'[^a-zA-Z0-9 ]', '', x)) 
df.reset_index(drop=True, inplace=True)
df = df.dropna(subset=['NAME'],inplace=True)
```

```python
df['NAME'].isnull().sum()
```

```python
df.dropna(subset=['NAME'],inplace=True)
```

```python
df['NAME'].isnull().sum()
```

```python
# 3.Host id column
df['host id'] = pd.to_numeric(df['host id'], errors='coerce')
df=df.dropna(subset=["host id"])
df.reset_index(drop=True, inplace=True)
```

```python
# 4.Host name column
df["host name"].astype(str).str.strip()
df['host name'].replace('', pd.NA, inplace=True)
```

```python
df.dropna(subset=["host name"],inplace=True)
```

```python
# 5. Neighbourhood column:
df.dropna(subset=["neighbourhood"],inplace=True)
df['neighbourhood'] = df['neighbourhood'].astype(str).str.strip().str.title()
```

```python
df['neighbourhood'].value_counts()
```

```python
df.columns
```

```python
df
```

```python
# 6.Latitude and longitude:
df.dropna(subset=["lat","long"],inplace=True)
df[(df['lat'] >= -90) & (df['lat'] <= 90)]
df[(df['long'] >= -180) & (df['long'] <= 180)]
```

```python
df["lat"].value_counts()
```

```python
# 7.Clean country ,country code ,instant bookable,cancellation
df.columns
```

```python
df['country'].astype(str).str.strip().str.title().replace("",pd.NA,inplace=True)
df.dropna(subset=["country"])
```

```python
df['country code'].astype(str).str.strip().str.title().replace("",pd.NA,inplace=True)
df.dropna(subset=['country code'],inplace=True)
```

```python
df.columns
```

```python
df.info()
```

```python
# 8.room_type column:
df["room type"]=df["room type"].astype(str).str.strip().str.title()
df.dropna(subset=["room type"],inplace=True)
```

```python
df["room type"].isnull().sum()
```

```python
df["room type"].value_counts()
```

```python
df
```

```python
#9.Clean construction year column
df.dropna(subset=["Construction year"],inplace=True)
df["Construction year"]=df["Construction year"].astype(int)
```

```python
df["Construction year"].isnull().sum()
```

```python
#10.Price columns
df.dropna(subset=["price"],inplace=True)
df["price"]=df["price"].str.strip()
df["price"]=df["price"].str.replace("[^\d]","",regex=True)
```

```python
import numpy as np
df["price"]=df["price"].replace("",np.nan,regex=True)
df["price"].isnull().sum()
```

```python
# 11.Clean service fee column
df['service fee'].isnull().sum()
```

```python
df.dropna(subset=['service fee'],inplace=True)
```

```python
# 12. minimum nights column:
df['minimum nights']=pd.to_numeric(df['minimum nights'])
df.dropna(subset=['minimum nights'],inplace=True)
df[df['minimum nights']>0]
```

```python
# 13. number_of_reviews columns
df['number of reviews']=pd.to_numeric(df['number of reviews'])
df.dropna(subset=['number of reviews'],inplace=True)
df[df['number of reviews']>0]
```

```python
# 14. last_review
df['last review']=pd.to_datetime(df['last review'],errors='coerce')
df.dropna(subset=['last review'],inplace=True)
```

```python
# 15. availability_365 column
df['availability 365']=pd.to_numeric(df['availability 365'])
df.dropna(subset=['availability 365'],inplace=True)
df[(df['availability 365']>=0) & (df['availability 365']<=365)]
```

```python
# 16. reviews_per_month columns
df["reviews per month"]=pd.to_numeric(df["reviews per month"])
df.dropna(subset=["reviews per month"])
df[df["reviews per month"]>=0]
```

```python
df.columns
```

```python
df.drop(columns=['host_identity_verified','neighbourhood group', 'review rate number', 'calculated host listings count','house_rules', 'license'],inplace=True)
```
